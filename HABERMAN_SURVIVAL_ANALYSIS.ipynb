```python
alive=haberman.loc[haberman["Survival_Status"]==1]
dead=haberman.loc[haberman["Survival_Status"]==2]
```

```python
counts, bin_edges = np.histogram(alive['positive_lymph'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:], cdf)
plt.legend(['Pdf for the patients who survive more than 5 years',
            'Cdf for the patients who survive more than 5 years'])
plt.show()
```

```python
counts, bin_edges = np.histogram(dead['positive_lymph'], bins=10, density=True)

pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:], cdf)
plt.legend(['Pdf for the patients who dead Within 5 years',
            'Cdf for the patients who dead within 5 years'])
plt.show()
```

```python
print("Summary statistics for the people alive for more than 5 years")
alive.describe()
```

```python
print("Summary statistics for the people dead for more than 5 years")
dead.describe()
```

```python
print("median")
print(np.median(alive['positive_lymph']))
print(np.median(np.append(alive['positive_lymph'],50)))
print(np.median(dead['positive_lymph']))

print("Quantiles")
print(np.percentile(alive['positive_lymph'],np.arange(0,100,25)))
print(np.percentile(dead['positive_lymph'],np.arange(0,100,25)))

print('90th Percentile')
print(np.percentile(alive['positive_lymph'],90))
print(np.percentile(dead['positive_lymph'],90))

from statsmodels import robust
print("Median Absolute Deviations")
print(robust.mad(alive["positive_lymph"]))
print(robust.mad(dead["positive_lymph"]))
```
